```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
plt.rcParams['figure.dpi'] = 300
%matplotlib inline
```

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
sigmas = 5
x = np.linspace(mu - sigmas*sigma, mu + sigmas*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
```

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
sigmas = 5
x = np.linspace(mu - sigmas*sigma, mu + sigmas*sigma, 1000)

variance = 1
sigma = math.sqrt(variance)
plt.plot(x, stats.norm.pdf(x, mu, sigma), label=f"$\mu={mu}, \sigma^2={variance}$")
variance = 1.5
sigma = math.sqrt(variance)
plt.plot(x, stats.norm.pdf(x, mu, sigma), label=f"$\mu={mu}, \sigma^2={variance}$")
variance = 0.5
sigma = math.sqrt(variance)
plt.plot(x, stats.norm.pdf(x, mu, sigma), label=f"$\mu={mu}, \sigma^2={variance}$")
plt.legend()

```
